6.2.1 결측치 처리
주어진 데이터 셋에 데이터가 존재하지 않으면 그것을 missing value라고 부른다. 이때 이것을 미리 처리하는것이 중요하다. 

numpy : 수학 연산
pandas : 데이터 프레임을 생성, 다루기 위한 라이브러리



```python
df = pd.DataFrame([
    [42,'male',12,'reading','class2'],
    [35,'unknown',3,'cooking','class1'],
    [1000,'female',7,'cycling', 'class3'],
    [1000, 'unknown',21,'unknown','unknown']])

df.columns = ['age','gender','month_birth','hobby','target']
df
```

```python
df['age'].unique()
df['gender'].unique()
df['month_birth'].unique()
df['hobby'].unique()
df['target'].unique()
```

```python
df.loc[df['age']>150, ['age']]=np.nan   ## 결측치를 NAN으로 바꾸는 과정
df.loc[df['gender']=='unknown',['gender']]=np.nan
df.loc[df['month_birth']>12, ['month_birth']] = np.nan
df.loc[df['hobby']=='unknown',['hobby']] = np.nan
df.loc[df['target']=='unknown',['target']] = np.nan

df
```

```python
#column에 있는 결측치 개수 확인

df.isnull().sum()

#결측치를 포함한 row 삭제
df2 = df.dropna(axis=0)
df2

```

```python
#결측치를 포함한 column 삭제
df3 = df.dropna(axis=1)
df3
```

```python
#모든 값이 결측치인 행 삭제

df4 = df.dropna(how = 'all')
df4
```

```python
#값이 2개 미만인 행 삭제

df5 = df.dropna(thresh=2)
df5
```

```python
#특정 column 에 결측치가 있는 경우 row 삭제

df6 = df.dropna(subset = ['gender'])
df6
```
